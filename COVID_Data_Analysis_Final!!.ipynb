<p style="font-family: 'Helvetica', sans-serif; font-size: 14px; color: #555;">

<p style="font-family: 'Helvetica', sans-serif; font-size: 14px; color: #555;">
    Consistency in data formats was achieved for accurate analysis.

<h2 style="font-family: 'Helvetica', sans-serif; font-size: 18px; color: #333;">
    <b>Handling Inconsistent Data</b>
</h2>
<p style="font-family: 'Helvetica', sans-serif; font-size: 14px; color: #555;">
    Inconsistencies in the dataset were identified and corrected for uniformity.
</p>
<ul style="font-family: 'Helvetica', sans-serif; font-size: 14px; color: #555; line-height: 1.6;">
    <li><b>Unique Values Check</b>: Unique entries in categorical columns were analyzed.</li>
    <li><b>Correction of Inconsistent Values</b>: Typos and variations in state names were corrected.</li>
    <li><b>Normalization of Text Data</b>: Text data was standardized to ensure case consistency.</li>
</ul>

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\covid_19_india.csv')

# 1. Handling Missing Values

# Detecting missing values
missing_values = df.isnull().sum()  # Check missing values in each column
print("Missing Values in Each Column:\n", missing_values)

# Dropping missing values
# Drop rows with any missing values
df_cleaned = df.dropna()

# Drop columns with many missing values
# df_cleaned = df.dropna(axis=1)  # Uncomment if you want to drop columns instead

# Filling missing values
# Fill with mean or median
df['Confirmed'] = df['Confirmed'].fillna(df['Confirmed'].mean())
df['Deaths'] = df['Deaths'].fillna(df['Deaths'].median())

# Forward fill
df.fillna(method='ffill', inplace=True)

# Using NumPy to replace missing values with a specific number (e.g., 0)
df['Confirmed'] = df['Confirmed'].replace(np.nan, 0)

# 2. Data Cleaning

# Removing Duplicates
duplicate_rows = df[df.duplicated()]  # Check for duplicates
print("Duplicate Rows:\n", duplicate_rows)

df_cleaned = df.drop_duplicates()  # Removes duplicates

# Standardizing Formats (e.g., Dates)
# Convert date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Remove rows with invalid dates
df_cleaned = df.dropna(subset=['Date'])

# 3. Handling Outliers
# Using IQR to detect outliers
Q1 = df['Confirmed'].quantile(0.25)
Q3 = df['Confirmed'].quantile(0.75)
IQR = Q3 - Q1

# Filtering out rows with outliers
df_cleaned = df[~((df['Confirmed'] < (Q1 - 1.5 * IQR)) | (df['Confirmed'] > (Q3 + 1.5 * IQR)))]

# Cap values to remove extreme outliers
df['Confirmed'] = df['Confirmed'].clip(lower=0, upper=df['Confirmed'].quantile(0.99))

# 4. Handling Inconsistent Data
# Check for unique values in the 'State' column
unique_states = df['State/UnionTerritory'].unique()
print("Unique States:\n", unique_states)

# Correct inconsistent values (example)
df['State/UnionTerritory'] = df['State/UnionTerritory'].replace({'Andhra Pardesh': 'Andhra Pradesh'})

# Normalizing Text Data (Case Sensitivity)
# Convert all state names to lowercase
df['State/UnionTerritory'] = df['State/UnionTerritory'].str.lower()

# Display cleaned data
print("Cleaned DataFrame:\n", df_cleaned.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tkinter as tk
from tkinter import ttk
from datetime import datetime

# Load the datasets
covid_df = pd.read_csv('C:\\Users\\Admin\\Downloads\\covid_19_india.csv')
vaccine_df = pd.read_csv('C:\\Users\\Admin\\Downloads\\covid_vaccine_statewise.csv')

# Data cleaning and preprocessing
def preprocess_data():
    global covid_df, vaccine_df
    
    # Drop unnecessary columns if they exist
    columns_to_drop = ["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"]
    covid_df.drop(columns=[col for col in columns_to_drop if col in covid_df.columns], inplace=True, axis=1)
    
    # Continue with data cleaning
    covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
    covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
    
    # Cleaning vaccination data
    vaccine_df.rename(columns={'Total Individuals Vaccinated': 'Total'}, inplace=True)
    vaccine_df = vaccine_df[vaccine_df.State != 'India']

preprocess_data()

# Function to plot active cases
def plot_active_cases(state=None, start_date=None, end_date=None):
    filtered_df = covid_df
    if state:
        filtered_df = filtered_df[filtered_df['State/UnionTerritory'] == state]
    if start_date:
        filtered_df = filtered_df[filtered_df['Date'] >= start_date]
    if end_date:
        filtered_df = filtered_df[filtered_df['Date'] <= end_date]
    
    top_10_active_cases = filtered_df.groupby(by='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by=['Active_cases'], ascending=False).reset_index()
    plt.figure(figsize=(16, 9))
    sns.barplot(data=top_10_active_cases.iloc[:10], y="Active_cases", x="State/UnionTerritory", linewidth=2, edgecolor='red')
    plt.title("Top 10 States with Most Active Cases in India", size=25)
    plt.xlabel("States")
    plt.ylabel("Total Active Cases")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to plot deaths
def plot_deaths(state=None, start_date=None, end_date=None):
    filtered_df = covid_df
    if state:
        filtered_df = filtered_df[filtered_df['State/UnionTerritory'] == state]
    if start_date:
        filtered_df = filtered_df[filtered_df['Date'] >= start_date]
    if end_date:
        filtered_df = filtered_df[filtered_df['Date'] <= end_date]
        
    top_10_deaths = filtered_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by='Deaths', ascending=False).reset_index()
    plt.figure(figsize=(18, 5))
    sns.barplot(data=top_10_deaths.iloc[:10], y="Deaths", x="State/UnionTerritory", linewidth=2, edgecolor='black')
    plt.title("Top 10 States with Most Deaths", size=25)
    plt.xlabel("States")
    plt.ylabel("Total Death Cases")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to plot vaccinations
def plot_vaccination():
    max_vac = vaccine_df.groupby('State')['Total'].sum().to_frame('Total').sort_values('Total', ascending=False).head(5)
    plt.figure(figsize=(10, 5))
    sns.barplot(data=max_vac, y=max_vac.Total, x=max_vac.index, linewidth=2, edgecolor='red')
    plt.title("Top 5 Vaccinated States In India", size=25)
    plt.xlabel("States")
    plt.ylabel("Total Vaccinations")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Main GUI Class
class COVID19AnalysisApp:
    def __init__(self, root):
        self.root = root
        self.root.title("COVID-19 Data Analysis in India")

        # Title Label
        self.title_label = ttk.Label(root, text="COVID-19 Data Analysis", font=("Helvetica", 16))
        self.title_label.pack(pady=10)

        # State Input
        self.state_label = ttk.Label(root, text="Enter State (Leave blank for all states):")
        self.state_label.pack(pady=5)
        self.state_entry = ttk.Entry(root)
        self.state_entry.pack(pady=5)

        # Start Date Input
        self.start_date_label = ttk.Label(root, text="Enter Start Date (YYYY-MM-DD):")
        self.start_date_label.pack(pady=5)
        self.start_date_entry = ttk.Entry(root)
        self.start_date_entry.pack(pady=5)

        # End Date Input
        self.end_date_label = ttk.Label(root, text="Enter End Date (YYYY-MM-DD):")
        self.end_date_label.pack(pady=5)
        self.end_date_entry = ttk.Entry(root)
        self.end_date_entry.pack(pady=5)

        # Graph Type Selection
        self.graph_type_label = ttk.Label(root, text="Select Graph Type:")
        self.graph_type_label.pack(pady=5)
        self.graph_type_combobox = ttk.Combobox(root, values=["Active Cases", "Deaths", "Vaccinations"])
        self.graph_type_combobox.pack(pady=5)

        # Button to show graph
        self.show_graph_button = ttk.Button(root, text="Show Graph", command=self.show_graph)
        self.show_graph_button.pack(pady=10)

        # Exit Button
        self.exit_button = ttk.Button(root, text="Exit", command=root.quit)
        self.exit_button.pack(pady=10)

    def show_graph(self):
        state = self.state_entry.get().strip()
        start_date = self.start_date_entry.get().strip()
        end_date = self.end_date_entry.get().strip()
        graph_type = self.graph_type_combobox.get()

        if graph_type == "Active Cases":
            plot_active_cases(state if state else None, 
                              start_date if start_date else None, 
                              end_date if end_date else None)
        elif graph_type == "Deaths":
            plot_deaths(state if state else None, 
                        start_date if start_date else None, 
                        end_date if end_date else None)
        elif graph_type == "Vaccinations":
            plot_vaccination()

# Run the GUI
if __name__ == "__main__":
    root = tk.Tk()
    preprocess_data()  # Preprocess the data before launching the GUI
    app = COVID19AnalysisApp(root)
    root.mainloop()
```
